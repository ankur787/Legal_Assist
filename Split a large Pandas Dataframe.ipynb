```python
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
```

```python
server = 'DESKTOP-HOTS4AV'
database = 'CustomerDatabaseAus'
driver = 'SQL Server Native Client 11.0'
connection = f'mssql://@{server}/{database}?driver={driver}'

engine = create_engine(connection,pool_size=10, max_overflow=20)
conn = engine.connect()
```

```python
sql = """SELECT * FROM [CustomerDatabaseAus].[dbo].[CellphoneDatabase] where States like 'N%' """
```

```python
df = pd.read_sql(sql, conn)
```

```python
df.head()
```

```python
df['Cellphone'] = df['Cellphone'].astype(int)
df['Cellphone'] = df['Cellphone'].astype(str)
```

```python
df['Country'].value_counts()
```

```python
df['Country'] = 'WA'
```

```python
df['Country_Code'] = '61'
df['Cellphone'] = df['Country_Code']+df['Cellphone']
df.drop('Country_Code', axis= 1, inplace= True)
df['Cellphone'] = df['Cellphone'].astype('int64')
```

```python
df.head()
```

```python
df_split= np.array_split(df, 4)
```
