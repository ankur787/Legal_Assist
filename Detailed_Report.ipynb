```python
import pandas as pd
import numpy as np
from scipy.stats import mode
pd.set_option('display.max_columns', None)
```

```python
df1 = pd.read_excel('disposition.xlsx')
df1.columns = df1.columns.str.replace(' ', '')
df1 = df1.loc[:, 'Lead':'CrmField-3']
df1.drop(['AgentID','Sip', 'Prefix'],axis = 1, inplace = True)
df1.rename(columns={'CrmField-1':'Title','CrmField-2':'FirstName','CrmField-3':'LastName'}, inplace=True)
```

```python
p = []

for i in df1['PhoneNumber']:
    a =  str(i)
    a = a.split('.')[0]
    b = a [-9:]
    p.append(b)
    
df1['PhoneNumber'] = p

d = []

for i in df1['DIDNumber']:
    c =  str(i)
    c = c.split('.')[0]
    b = c [-9:]
    d.append(b)
    
df1['DIDNumber'] = d

df1['Start Time']= pd.to_datetime(df1['StartTime'])
df1['End Time']= pd.to_datetime(df1['EndTime'])

cht = []
for i in df1['CallHandlingTime']:
    try:
        i = str(i)
        i = i.replace(":", "")
        h = i[:2]
        m = i[2:4]
        s = i[4:]
        h = int(h)
        m = int(m)
        s = int(s)
        min = (h*3600)+(m*60)+(s*1)
#        min = round(min,3)
        cht.append(min)
    except:
        cht.append(0)

df1['CallHandlingTime'] = cht

tt = []
for i in df1['TalkTime']:
    
    i = str(i)
    i = i.replace(":", "")
    h = i[:2]
    m = i[2:4]
    s = i[4:]
    h = int(h)
    m = int(m)
    s = int(s)
    min = (h*3600)+(m*60)+(s*1)
#        min = round(min,3)
    tt.append(min)


df1['TalkTime'] = tt

wd = []
for i in df1['WrapupDuration']:
    try:
        i = str(i)
        i = i.replace(":", "")
        h = i[:2]
        m = i[2:4]
        s = i[4:]
        h = int(h)
        m = int(m)
        s = int(s)
        min = (h*3600)+(m*60)+(s*1)
#        min = round(min,3)
        wd.append(min)
    except:
        wd.append(0)

df1['WrapupDuration'] = wd

df1.rename(columns={'PhoneNumber': 'DialedNumber', 'Lead':'LeadID','Disposition':'AgentDisposition' }
           ,inplace = True)


k = []

for i in df1['TalkTime']:
    if i <=30:
        k.append('Worst')
    elif i > 30 and i <=60:
         k.append('Poor')
    elif i > 60 and i <=180:
         k.append('Fair')
    elif i > 180 and i <=300:
         k.append('Good')
    else:
        k.append('Excellent')
        
df1['CallQuality'] = k


df1['DialedNumber'] = df1['DialedNumber'].astype(int)
df1['DIDNumber'] = df1['DIDNumber'].astype(int)

t = []
for i in df1['DialedNumber']:
    if i <400000000 or i>500000000:
        t.append('Landline')
    else:
        t.append('Cellphone')

df1['Type'] = t
```

```python
df = pd.read_html('cdr.xls')
df=pd.concat(df)
df=df.iloc[:,:]

dispos_col= ['#','Campaign','LeadID','DIDNumber','QueueName','Prefix','DialedNumber','CallType','Date/Time',
             'SIP','AgentID','AgentName','Pri/Voip','RingTime','BillSec','TotalTalktime','CDRDisposition',
             'HangupCode','HangupDescription','HangupBy']

df.columns = dispos_col
df.drop(['#','Campaign','QueueName','Prefix','SIP','Pri/Voip'],axis=1, inplace=True)
```

```python
p = []

for i in df['DIDNumber']:
    a =  str(i)
    a = a.split('.')[0]
    b = a [-9:]
    p.append(b)
    
df['DIDNumber'] = p

d = []

for i in df['DialedNumber']:
    c =  str(i)
    c = c.split('.')[0]
    b = c [-9:]
    d.append(b)
    
df['DialedNumber'] = d

df['BillSec'] = df['BillSec'].astype(str)
df['BillSec'] = df['BillSec'].str.replace(".","")
df['BillSec'] = df['BillSec'].str.slice(0,8)

df['RingTime'] = df['RingTime'].astype(str)
df['RingTime'] = df['RingTime'].str.replace(".","")
df['RingTime'] = df['RingTime'].str.slice(0,8)

df['TotalTalktime'] = df['TotalTalktime'].astype(str)
df['TotalTalktime'] = df['TotalTalktime'].str.replace(".","")
df['TotalTalktime'] = df['TotalTalktime'].str.slice(0,8)

tt = []
for i in df['TotalTalktime']:
    i = str(i)
    i = i.replace(":", "")
    h = i[:2]
    m = i[2:4]
    s = i[4:6]
    h = int(h)
    m = int(m)
    s = int(s)
    min = (h*3600)+(m*60)+(s*1)
    min = round(min,3)
    tt.append(min)

df['TotalTalktime'] = tt

bs = []
for i in df['BillSec']:
    try:
        i = str(i)
        i = i.replace(":", "")
        h = i[:2]
        m = i[2:4]
        s = i[4:6]
        h = int(h)
        m = int(m)
        s = int(s)
        min = (h*3600)+(m*60)+(s*1)
        min = round(min,3)
        
        bs.append(min)
    except:
        bs.append(0)
        

df['BillSec'] = bs 


df['DialedNumber'] = df['DialedNumber'].astype(int)
t = []
for i in df['DialedNumber']:
    if i <400000000 or i>500000000:
        t.append('Telephone')
    else:
        t.append('Cellphone')

df['Type'] = t


df['AgentName'].replace('nan', '', inplace=True)
df['HangupBy'].replace('nan', '', inplace=True)

df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df['Date']= df['Date/Time'].dt.date
df['Time']= df['Date/Time'].dt.time

df.drop('Date/Time', axis= 1, inplace= True)
```
